## 1. LOAD DATA:

```python
clusters = np.load('cluster1_answers.npy')
data = np.load('cluster1_data_scaled.npy')

print(data[:10])
print("")
print(clusters[:10])
plt.rcParams['figure.figsize'] = (8,6)

#to color each cluster, use c=clusters (clusters referanced above)
plt.scatter(data[:,0],data[:,1],c=clusters)
plt.show()
```

## 2. SLIT DATA INTO TRAINING AND TESTING SETS

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, clusters, test_size=0.33)
```

## 3. LETS CHECK OUT THE RESULTS:

><b> PLOT THE RESULTING TRAINING DATA

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
plt.show()
```

><b>PLOT THE RESULTING TESTING DATA

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_test)
plt.show()
```

><b> NOW THE TWO COMBINED:

```python
#ORIGINAL DATA
plt.scatter(data[:,0],data[:,1],c=clusters)
plt.show()

#SPLIT AND RECOMBINED DATA
plt.rcParams['figure.figsize'] = (8,6)
plt.scatter(x_train[:,0],x_train[:,1],c=y_train)
plt.scatter(x_test[:,0],x_test[:,1],c=y_test)
plt.show()
```

><b>It looks like the data has not changed. Whew!
